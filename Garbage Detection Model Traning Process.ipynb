<h1> Code to Display Random Images with Sizes </h1>

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import os

# Set the path to the training images folder
image_folder = '/content/drive/My Drive/NEWMODEL/train/images'

# Function to load and preprocess an image given its file path
def load_image(file_path):
    image = tf.io.read_file(file_path)  # Read the image file
    image = tf.image.decode_jpeg(image, channels=3)  # Decode the image to RGB format (channels=3) for JPEG images
    image = tf.image.resize(image, [640, 640])  # Resize the image to 640x640 pixels
    image = image / 255.0  # Normalize the image pixel values to the range [0, 1]
    return image

# Create a list of image file paths from the specified folder
image_paths = [os.path.join(image_folder, fname) for fname in os.listdir(image_folder) if fname.endswith(('jpg', 'png'))]

# Create a TensorFlow dataset from the list of image file paths
dataset = tf.data.Dataset.from_tensor_slices(image_paths)

# Apply the 'load_image' function to each image file path in the dataset
dataset = dataset.map(load_image)

# Batch the dataset, loading 10 images at a time
dataset = dataset.batch(10)  # Load 10 images at a time
```

```python
# Plot random samples from the dataset
plt.figure(figsize=(10, 4))  # Set the figure size for the plot
for image_batch in dataset.take(1):  # Take 1 batch of images from the dataset
    for i in range(10):  # Iterate over the 10 images in the batch
        ax = plt.subplot(2, 5, i + 1)  # Create a subplot grid with 2 rows and 5 columns
        plt.imshow(image_batch[i].numpy())  # Convert the tensor to a numpy array for displaying
        plt.axis('off')  # Hide axis for cleaner visualization
plt.show()  # Display the plot with the random image samples
```

<h1> 8. Real-Time Detection </h1>
Use a webcam or video feed for real-time detection.

```python
import cv2
from ultralytics import YOLO

# Load trained YOLO model
model = YOLO('/content/drive/My Drive/NEWMODEL/runs/detect/garbage_detection/weights/best.pt')

# Open video file
video_path = '/content/drive/My Drive/NEWMODEL/video.mp4'
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define output video writer
output_path = '/content/drive/My Drive/NEWMODEL/output_video.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform detection (no GUI)
    results = model.predict(source=frame, conf=0.25, stream=True)

    # Annotate the frame with results and write it
    for r in results:
        annotated_frame = r.plot()  # draw bounding boxes
        out.write(annotated_frame)

# Clean up
cap.release()
out.release()
```
