```python
import cv2
from ultralytics import YOLO

# Load the trained model (best.pt)
model = YOLO('best1.pt')  # Replace with the correct path to your 'best.pt' model

# Start the webcam (0 for the default camera)
cap = cv2.VideoCapture(0)

# Set the video capture frame width and height (optional)
cap.set(3, 1280)  # Width
cap.set(4, 1280)  # Height

while True:
    # Read a frame from the webcam
    ret, frame = cap.read()
    if not ret:
        break
    frame = cv2.flip(frame, 1)

    # Perform inference on the frame
    results = model(frame)  # The model processes the frame

    # Render predictions (bounding boxes, labels, etc.)
    output_frame = results[0].plot() 

    # Display the output frame
    cv2.imshow('Real-Time Detection', output_frame)

    # Press 'q' to quit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close any OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

# Load trained YOLO model
model = YOLO('best1.pt')  # Use your trained model path here

# Open video file
video_path = 'video.mp4'  # Change this to your actual video path
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(3))  # Frame width
frame_height = int(cap.get(4))  # Frame height
fps = int(cap.get(cv2.CAP_PROP_FPS))  # Frames per second

# Define output video writer (optional: saves the detected video)
output_path = 'output_video.mp4'  # Path for saving the result video
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for saving the video
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Stop when video ends

    # Perform detection on the frame
    results = model.predict(source=frame)  # Get the detection results

    # Iterate through the detected objects
    for result in results.xyxy[0]:  # result contains bounding box details
        x1, y1, x2, y2, conf, cls = result  # Extract bounding box coordinates and class info
        label = results.names[int(cls)]  # Get the class label (e.g., 'person')
        confidence = conf.item()  # Get the confidence score

        # Draw the bounding box on the frame
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)

        # Add the label and confidence score near the bounding box
        cv2.putText(frame, f"{label} {confidence:.2f}", (int(x1), int(y1)-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Save the frame with detections to the output video
    out.write(frame)

    # Optionally, display the frame with the bounding boxes (for debugging or checking)
    cv2.imshow('Detected Frame', frame)

    # Press 'q' to quit early
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture and writer objects
cap.release()
out.release()

# Close all OpenCV windows
cv2.destroyAllWindows()
```
